```python
import pandas as pd
import numpy as np
import string
string.punctuation
```

```python
data=pd.read_csv('lf.csv')
```

```python
pd.set_option('display.max_colwidth', -1)
data = data[['review_text']]
data.head()
```

```python
# data['review_text'].value_counts()
```

```python
def remove_punctuation(text):
    punctuationfree = "".join([i for i in text if i not in string.punctuation])
    return punctuationfree


#storing the puntuation free text
data['clean_msg'] = data['review_text'].apply(lambda x: remove_punctuation(x))
data.head()
```

```python
data['msg_lower']= data['clean_msg'].apply(lambda x: x.lower())
data
```

```python
import re


def tokenization(text):
    tokens = re.split('W+', text)
    return tokens


#applying function to the column
data['msg_tokenied'] = data['msg_lower'].apply(lambda x: tokenization(x))
data
```

```python
import nltk
#Stop words present in the library
stopwords = nltk.corpus.stopwords.words('english')
stopwords[0:10]
['i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', "you're"]
```

```python
#defining the function to remove stopwords from tokenized text
def remove_stopwords(text):
    output = [i for i in text if i not in stopwords]
    return output
```

```python
#applying the function
data['no_stopwords'] = data['msg_tokenied'].apply(
    lambda x: remove_stopwords(x))
data
```

```python
#importing the Stemming function from nltk library
from nltk.stem.porter import PorterStemmer
#defining the object for stemming
porter_stemmer = PorterStemmer()
```

```python
#defining a function for stemming
def stemming(text):
    stem_text = [porter_stemmer.stem(word) for word in text]
    return stem_text


data['msg_stemmed'] = data['no_stopwords'].apply(lambda x: stemming(x))
data
```

```python
from nltk.stem import WordNetLemmatizer
import nltk
#defining the object for Lemmatization
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
data['msg_stemmed'].to_csv('file4.csv',index=True)
```
